```python
import matplotlib.pyplot as plt
```

```python
X = [0, 8, 15, 22]
Y = [32, 46.4, 59, 71.6]
```

```python
# Plot dot
plt.plot(X, Y, 'ro')
plt.show()
```

```python
# Plot dot with axis range
plt.plot(X, Y, 'ro')
plt.axis([0, 25, 0, 100])
plt.show()
```

```python
# Plot with default line
plt.plot(X, Y)
plt.axis([0, 25, 0, 100])
plt.show()
```
